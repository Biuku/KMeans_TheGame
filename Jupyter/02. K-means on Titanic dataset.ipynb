```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import numpy as np
import pandas as pd
```

```python
arr = np.load('../data/arr250.npy')
```

```python
arr[:5]
```

## 1. Plot the clusters

```python
a = arr[:,0]
b = arr[:,1]
labels = arr[:,2]
```

```python
fig = plt.figure(figsize = (12, 5))

plt.scatter(a, b, c=labels, cmap='rainbow')
```

# 2. K-means

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)

km.fit(arr[:,:2])
```

```python
km.cluster_centers_
```

#### The predicted labels

```python
km.labels_
```

```python
arr[:,2]
```

## 3. Plot performance

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(12, 5))

ax1.set_title('Actual Titanic survivors')
ax1.scatter(a, b, c=arr[:,2], cmap='rainbow')

ax2.set_title('Predicted survivors')
ax2.scatter(a, b, c=km.labels_, cmap='rainbow')
```

## Interesting!

## 4. Performance

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(arr[:,2], km.labels_))
```

```python
print(classification_report(arr[:,2], km.labels_))
```
