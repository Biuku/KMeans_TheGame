```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

## 1. Make a dataset of 'blob' clusters

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(200, 2, 4, 1.8, random_state = 101)
```

```python
len(data)
```

```python
data[0].shape
```

#### col 1 = the 'correct' labels

```python
data[1][:100]
```

```python
labels = data[1]
```

```python
a = data[0][:,0]
b = data[0][:,1]

plt.scatter(a, b, c=labels, cmap='rainbow')
```

## 2. K-means

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)

kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

#### The predicted labels

```python
kmeans.labels_
```

## 3. Plot performance

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))

ax1.set_title('Actual')
ax1.scatter(a, b, c=labels, cmap="rainbow")

ax2.set_title('K-means predicted')
ax2.scatter(a, b, c=kmeans.labels_, cmap="rainbow")
```
